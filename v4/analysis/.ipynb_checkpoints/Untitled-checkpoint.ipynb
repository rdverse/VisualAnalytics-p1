```python
personList = list()
for doc in data['documents']['document']:
    datum =list()
    try:
        datum = doc['Person']
        personList.append(datum)
    except:
        pass
    
# Create matries - person-person
#                - person-organization
```

```python
personList
```

```python
import pandas as pd
import itertools
c = [list(itertools.combinations(i,2)) for i in personList]
a = list(itertools.chain.from_iterable((i, i[::-1]) for c_ in c for i in c_))
df = pd.DataFrame(a)
pd.pivot_table(df, index=0, columns=1, aggfunc='size', fill_value=0)
```

```python
combList= list()
for doc in data['documents']['document']:
    list1 = list()
    list2 = list()
    
    try:
        datum1 = doc['Organization']
        datum2 = doc['Person']
        
        list1.extend(datum1)
        list2.extend(datum2)
        
        list2 = [[l] for l in list2]
        
#         print(list2[0]*len(list1))
#         print(list1)
        print(list(zip(list2[0]*len(list1), list1)))
        
        #print( np.hstack((np.array(list2[0]*len(list1)).reshape(-1,1), np.array(list11.reshape(-1,1)))))
        
        for l2 in list2:
            combList.append(list(zip(l2*len(list1), list1)))
        
    except:
        pass
    
```

```python

import numpy as np
pd.DataFrame(np.ravel(combList))
```
